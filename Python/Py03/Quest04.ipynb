```python
import math   #도구함에서 수학 명령어 가져오기

while True: # 입력값이 알맞을때('참'인경우 계속해서 반복해라)
    try:  #실행도중 에러가 발생할 수 있으므로 미리 대비하기위해 사용

        num1 = int(input("첫 번째 정수: "))  #정수(숫자)만 받기위해서 앞에 인트로 묶어줍니다
        num2 = int(input("두 번째 정수: "))
        calc = input("연산자(+, -, *, /, **): ") #연산자 로 입력할 기호를 튜플로 변경해서"( )" 기록해줍니다

        # 2. 계산 및 예외 처리
        if calc == '+': result = num1 + num2  # +부터 if로 시작 나머지는 이프가 아니고 - or *,/ 등 라면 elif 시작
        elif calc == '-': result = num1 - num2
        elif calc == '*': result = num1 * num2
        elif calc == '/':
            if num2 == 0: raise ZeroDivisionError("""나누는 값이 "0"일 수 없습니다.""")  #그중에서 만약 0으로 몫을 나눌때 오류값 설정, raise 를써서 except로 에러를 보냄(파이썬 강제종료 예방)
            result = num1 / num2
        elif calc == '**': result = math.pow(num1, num2) # math 모듈 사용
        else:
            print("지원하지 않는 연산자입니다.")    #위의 조건에 해당되지 않을때 즉 지정해준 연산자가 아닌 str같은 값을 입력했을때
            continue

        print(f"값: {result}")

    except ValueError: #  정수 입력 (잘못 입력하면 ValueError 발생 -> except로 이동)
        print("잘못된 입력입니다. 정수를 입력해주세요.")
        continue # 밸류 에러일때 반복문의 처음 wille<==로 다시 시작하시오
    except ZeroDivisionError as e: #굳이 as e를 쓰는이유 : 어떤 에러가 났는지 zeroDivisionError<-- 저장된 문구가  파이썬에서 출력되기때문
        print(e)
        continue

    # 3. 종료 여부 확인
    if input("계속하시겠습니까? (y/n): ") != 'y':  #마지막으로 y가 아닐때 while 반복함수를 끝내라 ==y 일때는 계속 반복하라.
        break
```
